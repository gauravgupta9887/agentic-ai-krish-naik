### Model Integration with OpenAI, Google Geimini and GROQ

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.agents import create_agent
from langchain_openai import ChatOpenAI

custom_llm = ChatOpenAI(
    model="gemma-local",
    api_key=os.getenv("CHAT_API_KEY"),
    base_url=os.getenv("CHAT_API_BASE_URL"),
    temperature=0,
    verbose=True
)
```

```python
custom_llm.invoke("Who was the first person to walk on the moon?")
```

```python
response = custom_llm.invoke("Why does parrot talk?")
response.content
```
